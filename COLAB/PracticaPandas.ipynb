### Parte 2: Limpieza y Preparación de Datos
1. Convertir el peso de Kilos a Libras.
2. Convertir Edad de Años a Meses.
3. Tratar con valores 0 en diabetes (si los hay): Identificar y manejar los valores 0 es en el conjunto de datos.

```python
# 1. CONVERTIMOS LOS KILOS EN LIBRAS
df['Peso'] = df['Peso'] * 2.20462
print("DF CON PESO EN LIBRAS")
df
```

```python
# 2. AÑOS A MESES
df['Edad'] = round(df['Edad'] * 12)
print("DF CON EDAD EN MESES")
df
```

```python
# 3. REMPLAZAMOS EL 0 DE DIABETES EN NaN
df['Diabetes'] = df['Diabetes'].replace(0, pd.NA)
print("DF CON DIABETES EN NA")
df
```

```python
# Revertimos los datos nulos a 0
df['Diabetes'] = df['Diabetes'].fillna(0)
df
```

```python
# Ponemos los valores 0 a nulos
df['Diabetes'] = df['Diabetes'].replace(0, 'null')
print("DF CON DIABETES EN NULL")
df
```

### Parte 3: Análisis Exploratorio de Datos
1. Agrupar datos por Diabetes y calcular el peso medio para cada Tipo de diabetes.
2. Visualización:
  1. Crear histogramas de las distribuciones de edad y peso.
  2. Crear un diagrama de caja (boxplot) para comparar la distribución del peso entre géneros.
3. Filtrar datos para un rango de peso específico y calcular estadísticas descriptivas.


```python
# 1. AGRUPAR DATOS POR TIPO DE DIABETES
pesoportipodiabetes = df.groupby('Diabetes')['Peso'].mean()
print("PESO MEDIO POR TIPO DE DIABETES")
pesoportipodiabetes
```

```python
# 2.1 HISTOGRAMA PESO Y EDAD

# Crear los histogramas en una sola fila (2 columnas)
plt.figure(figsize=(14, 6))  # Tamaño general de la figura

# Histograma del peso
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, gráfico en la posición 1
sns.histplot(df['Peso'], color='blue', bins=15)  # KDE para la curva de densidad
plt.title('Distribución de Peso')
plt.xlabel('Peso (kg)')
plt.ylabel('Frecuencia')

# Histograma de la edad
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, gráfico en la posición 2
sns.histplot(df['Edad'], color='green', bins=15)  # KDE para la curva de densidad
plt.title('Distribución de Edad')
plt.xlabel('Edad (años)')
plt.ylabel('Frecuencia')

# Mostrar los gráficos
plt.tight_layout()  # Ajusta automáticamente los espacios entre los gráficos
plt.show()
```

```python
# BOXPLOT DE PESO VS. GENERO

# Crear un boxplot para comparar el peso entre géneros
plt.figure(figsize=(6, 4))
sns.boxplot(x='Genero', y='Peso', data=df, palette='Set2')

# Personalizar el gráfico
plt.title('Distribución del Peso entre Géneros')
plt.xlabel('Género')
plt.ylabel('Peso (kg)')

# Mostrar el gráfico
plt.show()
```

```python
# 3. FILTRAMOS DATOS POR EDAD
# PASAMOS DE LIBRAS A KG Y EDAD EN AÑOS
df['Peso'] = df['Peso'] / 2.20462
df['Edad'] = df['Edad'] / 12

print("DF CON PESO EN KG Y EDAD EN AÑOS")
df
```

```python
# Filtrar los datos por un rango específico de edad (por ejemplo, entre 30 y 50 años)
rango_edad = df[(df['Edad'] >= 30) & (df['Edad'] <= 40)]

# Calcular estadísticas descriptivas para las columnas numéricas (por ejemplo, peso_kg)
estadisticas = rango_edad.describe()

# Mostrar las estadísticas descriptivas
print("\nEstadísticas descriptivas para el rango de edad entre 30 y 50 años:")
print(estadisticas)
```

### Parte 4: Operaciones Avanzadas
1. Calcular los percentiles del peso para cada género.
2. Crear una nueva columna que clasifique a los individuos como 'Bajo Peso','Peso Normal', 'Sobrepeso' o 'Obesidad' basado en su peso


```python
# CALCULO DE PERCENTILES PARA CADA GENERO
# Agrupar los datos por género y calcular los percentiles del peso
percentiles = df.groupby('Genero')['Peso'].quantile([0.25, 0.50, 0.75])

# Mostrar los percentiles
print("\nPercentiles del peso por género:")
print(percentiles)
```

```python
# Crear columna con clasificacion de pesos segun obesidad

df['Obesidad'] = np.where(df['Peso'] < 50, 'Bajo Peso',
                          np.where(df['Peso'] < 70, 'Peso Normal',
                          np.where(df['Peso'] < 90, 'Sobrepeso',
                                   'Obesidad')))

# Mostrar los resultados con la nueva columna
print("\nDatos con clasificación de peso:")
print(df[['Nombre', 'Genero', 'Peso', 'Diabetes', 'Obesidad']])
```
